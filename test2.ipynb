```python
x_data = np.array([15,87,138,178,206,228,246,260,274,283,291,300,307])
y_data = np.array([40,50,60,70,80,90,100,110,120,130,140,150,160])
```

```python
plt.errorbar(x_data, y_data, ls='', capsize=4, marker='.', ms=5, mec='black')
```

```python
def fit_func(x, a, b,):
    return a*np.exp(-b*x)
popt, pcov = curve_fit(fit_func, x_data, y_data)
perr = np.sqrt(np.diag(pcov))
```

```python
x_model = np.linspace(min(x_data), max(x_data), 100)
y_model = fit_func(x_model, *popt)
#print(x_model)
print(y_model)
```
