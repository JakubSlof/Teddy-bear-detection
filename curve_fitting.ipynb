```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
from scipy.optimize import curve_fit
```

```python
x_data = np.array([15,87,138,178,206,228,246,260,274,283,291,300,307,312])
y_data = np.array([40,50,60,70,80,90,100,110,120,130,140,150,160,170])
```

```python
plt.scatter(x_data,y_data)
plt.show()
```

```python
def model(x,a,b,c):
    return a*(x-b)**2 + c

popt, pcov = curve_fit(model, x_data,y_data)
```

```python
print(popt)
a,b,c = popt
print(a)
print(b)
print(c)
```

```python
pcov
```

```python

x_model = np.linspace(min(x_data),max(x_data),100)
y_model = model(x_model,*popt)
```

```python
plt.scatter(x_data,y_data)
plt.plot(x_model,y_model,color='r')
plt.show()
```
